Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of
features. What is the minimum number of principal components required to explain 90% of the variance in
the data?

Answer

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
wine_data_std=scaler.fit_transform(df)
```

```python
from sklearn.decomposition import PCA
pca=PCA()
pca.fit(wine_data_std)
```

We can examine the explained variance ratio of each principal component using the explained_variance_ratio_attribute of the PCA object:
    

```python
explained_variance=pca.explained_variance_ratio_
explained_variance
```

```python
explained_variance_percentage=[each_variance*100 for each_variance in explained_variance]
explained_variance_percentage
```

from the explained variance ratios,we can see the first principal component explains the highest amount of variance(26.0%),followed by the second and third principal components(18.6%and14.0%,respectively).
To determine the minimum number of principal components required to explain 90% of the variance in the data,we can calculate the cumulative explained variance ratio and find the index of the first principal component that exceeda 90%.

```python
import numpy as np
cumulative_variance=np.cumsum(explained_variance)
n_components=np.argmax(cumulative_variance>=0)+1
print(f"minimum number of principal components required to explain 90%of variance:{n_components}")
```
